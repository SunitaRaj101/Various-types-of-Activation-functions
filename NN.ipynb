```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def LinAct(X):
    return X
X=np.linspace(-10,10)
plt.plot(X,LinAct(X))
plt.show()
```

```python
def tanhyperbolic(X):
    return np.tanh(X)
X=np.linspace(-10,10)
plt.plot(X,tanhyperbolic(X))
plt.show()
    
```

```python
def sigmoidfn(X):
    return 1/(1+np.exp(-X))
X=np.linspace(-10,10)
plt.plot(X,sigmoidfn(X))
plt.show()
```

```python
def Relufn(X):
        return max(0,X)
iss=[x for x in range(-10,10)]
print(iss)

oss=[Relufn(X)  for X in iss]
print(oss)
plt.plot(iss,oss)
plt.show()
```

```python
def bipolarsigmoid(X):
    return (2/(1+np.exp(-X)))-1
X=np.linspace(-10,10)
plt.plot(X,bipolarsigmoid(X))
plt.show()
```

```python
#binary step func
print(np.heaviside([-5,5],10))
def binarystepfn(X):
    return np.heaviside(X,1)
X=np.linspace(-10,10)
plt.plot(X,binarystepfn(X))
plt.show()
```

```python
def summationfn(X):
    return np.exp(X)/(np.exp(sum(X)))
X=np.linspace(-10,10)
plt.plot(X,summationfn(X))
plt.show()
```
